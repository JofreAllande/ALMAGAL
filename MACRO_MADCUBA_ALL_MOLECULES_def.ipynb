```python
import os
import shutil

def delete_macros(carpeta_actual):
    directorio_original = os.getcwd()  # Guarda el directorio de trabajo original
    try:
        os.chdir(carpeta_actual)  # Cambia al directorio actual
        subcarpetas = [subcarpeta for subcarpeta in os.listdir(carpeta_actual) if os.path.isdir(os.path.join(carpeta_actual, subcarpeta))]

        for subcarpeta in subcarpetas:
            #if subcarpeta == '101899':
                print(subcarpeta)
                try:
                    # Define la ruta donde buscar los archivos
                    ruta = os.path.join(carpeta_actual, subcarpeta, 'analysis/spectra/7MTM2TM1_newvel')

                    # Verifica si la ruta existe
                    if os.path.exists(ruta):
                        os.chdir(ruta)  # Cambia al directorio especificado
                        # Lista los archivos en la carpeta y elimina los que empiezan con "MAD_ALMAGAL"
                        archivos = os.listdir(ruta)
                        #print(archivos)
                        archivos_a_borrar = [archivo for archivo in archivos if archivo.startswith('MACRO_SEARCH')]
                        #print(archivos_a_borrar)
                        for archivo in archivos_a_borrar:
                            os.remove(os.path.join(ruta, archivo))  # Elimina el archivo
                            print(f"Archivo eliminado: {archivo}")

                        os.chdir(carpeta_actual)  # Regresa a la carpeta principal
                    else:
                        print(f"No existe el directorio: {ruta}")

                except Exception as e:
                    print(f"Error procesando {subcarpeta}: {e}")

    finally:
        os.chdir(directorio_original)  # Regresa al directorio original al finalizar

# Example usage:
delete_macros(sources_path)
```
